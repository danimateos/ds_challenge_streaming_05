````python
pd.options.display.max_columns = None
bookings_sample = pd.read_csv("bookings.csv.bz2", sep="^", nrows=20000)
bookings_sample.head()
````

````python
bookings_sample = pd.read_csv("bookings.csv.bz2", sep="^", nrows=20000)
bookings_sample = bookings_sample[['arr_port', 'pax', 'year']]
bookings_2013 = bookings_sample[bookings_sample['year']==2013]
arrivals = bookings_2013.groupby('arr_port').sum()[['pax']]
arrivals.sort_values(by="pax", ascending=False).head(10)
````

````python
bookings_sample = pd.read_csv("bookings.csv.bz2", sep="^", nrows=100000, chunksize=10000)

partial_results = []

for chunk in bookings_sample:
    chunk = chunk[['arr_port', 'pax', 'year']]
    bookings_2013 = chunk[chunk['year']==2013]
    arrivals = bookings_2013.groupby('arr_port').sum()[['pax']]
    
    partial_results.append(arrivals)
    
total = pd.concat(partial_results).groupby('arr_port').sum()[['pax']]
top_10 = total.sort_values(by="pax", ascending=False).head(10)
top_10
````

````python
%%time

bookings = pd.read_csv("bookings.csv.bz2", sep="^", chunksize=100000)

partial_results = []

for chunk in bookings:
    chunk = chunk[['arr_port', 'pax', 'year']]
    bookings_2013 = chunk[chunk['year']==2013]
    arrivals = bookings_2013.groupby('arr_port').sum()[['pax']]
    
    partial_results.append(arrivals)
    
total = pd.concat(partial_results).groupby('arr_port').sum()[['pax']]
top_10 = total.sort_values(by="pax", ascending=False).head(10)
top_10
````
